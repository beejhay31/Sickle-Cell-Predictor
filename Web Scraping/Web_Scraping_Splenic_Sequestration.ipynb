```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
import requests
url = "https://wikem.org/wiki/Splenic_sequestration"
# makes a request to the web page and gets its HTML
r = requests.get(url)
# stores the HTML page in 'soup', a BeautifulSoup object
soup = BeautifulSoup(r.content,"lxml")
```

```python
print(soup.prettify())
```

```python
df2 = pd.DataFrame(columns =['Subheading', 'List1'])
#heading = []
subheading2 = []
lists2 = []
#lists2 = []
flag1 = 0
for link2 in soup.find_all("h2"):
  if flag1 == 0:
    subheading2.append(link2.get_text())
    next_sib2 = link2.find_next_sibling(['ul', 'h2'])
    if next_sib2 is not None:
      print(next_sib2.get_text())
      text2 = []
      while next_sib2.name != 'h2':
        text2.append(next_sib2.get_text())
        next_sib2 = next_sib2.find_next_sibling(['ul','h2'])
      if next_sib2 is None :
        flag1 = 1
        break
      lists2.append(' '.join(text2))
  else:
    break
df2['Subheading'] = subheading2
df2 = df2.drop(index=[0,9])
df2["List1"] = lists2
```
