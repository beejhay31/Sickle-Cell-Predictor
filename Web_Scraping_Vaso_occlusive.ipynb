```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
import requests
url = "https://wikem.org/wiki/Vaso-occlusive_pain_crisis"
# makes a request to the web page and gets its HTML
r = requests.get(url)
# stores the HTML page in 'soup', a BeautifulSoup object
soup = BeautifulSoup(r.content,"lxml")
```

```python
print(soup.prettify())
```

```python
df = pd.DataFrame(columns =['Subheading', 'List1'])
#heading = []
subheading = []
lists1 = []
#lists2 = []
flag = 0
for link1 in soup.find_all('h3'):
  if flag == 0:
    subheading.append(link1.get_text())
    #heading.append(link2.get_text())
    next_sib1 = link1.find_next_sibling(['ul', 'h3'])
    #next_sib2 = link2.find_next_sibling(['ul', 'h2'])
    print(next_sib1.get_text())
    #print(next_sib2.get_text())
    text1 = []
    #text2 = []
        # loop to go through all paragraphs that come after a h3 tag
    while next_sib1.name != 'h3':
      #while next_sib.name != 'h2':
      text1.append(next_sib1.get_text())
      #text2.append(next_sib2.get_text())
      next_sib1 = next_sib1.find_next_sibling(['ul', 'h3'])
      #next_sib2 = next_sib2.find_next_sibling(['ul','h2'])
      if next_sib1 is None :
        #if next_sib2 is None :
        flag = 1
        break
    lists1.append(' '.join(text1))
    #lists2.append(' '.join(text2))
  else:
    break
df['Subheading'] = subheading
#df['Heading'] = heading
df['List1'] = lists1
#df['List2'] = lists2
```
